```python
# === Polynomial Regression Degree 2 ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Baca dataset
df = pd.read_csv("/content/sample_data/dataset_cabai_baru.csv")

X = df['Minggu'].values.reshape(-1, 1)
y = df['Harga_Cabai'].values
```

```python
# Degree 2
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)

model2 = LinearRegression()
model2.fit(X_poly2, y)
```

```python
# Cetak Koefisien
print("=== Koefisien Degree 2 ===")
for i, coef in enumerate(model2.coef_):
    print(f"θ{i} : {coef}")
print("Intercept (θ0) =", model2.intercept_)
```

```python
# Grafik
X_range = np.linspace(1, 20, 300).reshape(-1, 1)
X_range_poly2 = poly2.transform(X_range)
y_pred2 = model2.predict(X_range_poly2)

plt.scatter(X, y, label="Data Asli")
plt.plot(X_range, y_pred2, color='red', label="Polynomial Degree 2")
plt.title("Polynomial Regression Degree 2")
plt.xlabel("Minggu")
plt.ylabel("Harga Cabai (Rp)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
minggu_baru = [[21]]
prediksi_2 = model2.predict(poly2.transform(minggu_baru))
print("Prediksi minggu ke-21 (Degree 2):", prediksi_2[0])
```
