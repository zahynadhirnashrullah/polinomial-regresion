```python
# === Polynomial Regression Degree 2 ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Baca dataset
df = pd.read_csv("/content/sample_data/dataset_cabai_baru.csv")

X = df['Minggu'].values.reshape(-1, 1)
y = df['Harga_Cabai'].values
```

```python
# Degree 7
poly7 = PolynomialFeatures(degree=7)
X_poly7 = poly7.fit_transform(X)

model7 = LinearRegression()
model7.fit(X_poly7, y)
```

```python
# Cetak Koefisien
print("=== Koefisien Degree 7 ===")
for i, coef in enumerate(model7.coef_):
    print(f"θ{i} : {coef}")
print("Intercept (θ0) =", model7.intercept_)
```

```python
# Grafik
X_range = np.linspace(1, 20, 300).reshape(-1, 1)
X_range_poly7 = poly7.transform(X_range)
y_pred7 = model7.predict(X_range_poly7)

plt.scatter(X, y, label="Data Asli")
plt.plot(X_range, y_pred7, color='green', label="Polynomial Degree 7")
plt.title("Polynomial Regression Degree 7")
plt.xlabel("Minggu")
plt.ylabel("Harga Cabai (Rp)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# === Prediksi Minggu ke-21 (Degree 7) ===
minggu_baru = [[21]]
prediksi_7 = model7.predict(poly7.transform(minggu_baru))
print("\nPrediksi harga cabai minggu ke-21 (Degree 7):", prediksi_7[0])
```
